```python
classical_tracks = pd.concat([bach_tracks, barber_tracks, bartok_tracks, beethoven_tracks, borodin_tracks, brahms_tracks, britten_tracks, bruch_tracks, debussy_tracks, delius_tracks, dvorak_tracks, elgar_tracks, faure_tracks, glazunov_tracks, handel_tracks, hayden_tracks, khachaturian_tracks, liszt_tracks, mendelssohn_tracks, mozart_tracks, mussorgsky_tracks, poulenc_tracks, prokofiev_tracks, puccini_tracks, rachmaninoff_tracks, ravel_tracks, respighi_tracks, strauss_tracks, korsakov_tracks, rossini_tracks, saens_tracks, satie_tracks, schubert_tracks, schumann_tracks, schostakovich_tracks, sibelius_tracks, smetana_tracks, tchaikovsky_tracks, williams_tracks, verdi_tracks, vivaldi_tracks, wagner_tracks], axis=0)
#classical_tracks
```

```python
classical_tracks = classical_tracks[["track.id", "track.name", "track.popularity"]]
classical_tracks = classical_tracks.rename(columns={"track.id": "track_id", "track.name": "track_name", "track.popularity": "track_popularity"})
#classical_tracks
```

```python
chunks = [(i, i+100) for i in range(0, len(classical_tracks), 100)]
audio_features = []
for chunk in chunks:
    id_list_100 = classical_tracks["track_id"][chunk[0]:chunk[1]]
    audio_features = audio_features + sp.audio_features(id_list_100)
    sleep(randint(1,3000)/1000)
#audio_features
```

```python
audio_features = json_normalize(audio_features)
#audio_features
```

```python
classical_tracks_audio_features = pd.merge(left=classical_tracks,
                                           right=audio_features,
                                           how="inner",
                                           left_on="track_id",
                                           right_on="id")
#classical_tracks_audio_features
```

```python
classical_tracks_audio_features = classical_tracks_audio_features.drop(["track_id", "type", "uri", "analysis_url"], axis=1)
#classical_tracks_audio_features
```

```python
classical_tracks_audio_features_num = classical_tracks_audio_features.drop(["track_name", "id", "track_href"], axis=1)
```

```python
X_standard_scaled = StandardScaler().fit_transform(classical_tracks_audio_features_num)
#pd.DataFrame(X_standard_scaled, columns=classical_tracks_audio_features_num.columns).head()
```

```python
kmeans = KMeans(n_clusters=10, random_state=13)
kmeans.fit(X_standard_scaled)
```

```python
clusters = kmeans.predict(X_standard_scaled)
pd.Series(clusters).value_counts().sort_index()
```

```python
plt.figure(figsize=(5, 5))
plt.scatter(X_standard_scaled[:, 0], X_standard_scaled[:, 10], c=clusters, alpha=0.5)
plt.show()
```

```python
%%time
kmeans = KMeans(n_clusters=13,
                init="random",
                n_init= 1,
                max_iter=13,
                tol=0,
                random_state=13)
kmeans.fit(X_standard_scaled)
print(kmeans.inertia_)
```

```python
K = range(2, 20)
inertia = []

for k in K:
    kmeans = KMeans(n_clusters=k,
                    random_state=13)
    kmeans.fit(X_standard_scaled)
    inertia.append(kmeans.inertia_)

%matplotlib inline

plt.figure(figsize=(16,8))
plt.plot(K, inertia, "bx-")
plt.xlabel("k")
plt.ylabel("inertia")
plt.xticks(np.arange(min(K), max(K)+1, 1.0))
plt.title("Elbow Method showing the optimal k")
```

```python
K = range(2, 20)
silhouette = []

for k in K:
    kmeans = KMeans(n_clusters=k,
                    random_state=13)
    kmeans.fit(X_standard_scaled)
    silhouette.append(silhouette_score(X_standard_scaled, kmeans.predict(X_standard_scaled)))


plt.figure(figsize=(16,8))
plt.plot(K, silhouette, "bx-")
plt.xlabel("k")
plt.ylabel("silhouette score")
plt.xticks(np.arange(min(K), max(K)+1, 1.0))
plt.title("Slhouette score showing the optimal k")
```

```python
agglomerative = cluster.AgglomerativeClustering(n_clusters=10, linkage="average")
pred = agglomerative.fit_predict(X_standard_scaled)

plt.figure(figsize=(5, 5))
plt.scatter(X_standard_scaled[:, 0], X_standard_scaled[:, 10], c=pred,alpha=0.5)
plt.show()
```
